```python
import numpy as np
from numpy import sin, cos
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def pen(n):

    g=1
    l = 1

    t= np.zeros([n+1, 1]) #time
    theta = np.zeros([n+1, 1])
    v = np.zeros([n+1, 1])

    theta[0] = 0.0
    t[0] = 0.0
    v[0] = 1.0

    max_theta = 0

    tfinal = 6 * np.pi

    # dt = (tfinal -t[0]) / float(n)
    dt = (tfinal-t[0])/float(n)

    for i in range(n):
        theta_mid = theta[i] + 0.5 * dt * v[i]
        y_mid = v[i] + 0.5 * dt *(-g/l * np.sin(theta[i]))
        t[i+1] = t[i] + dt
        theta[i + 1] = theta[i] + dt * (y_mid)
        v[i+1] = v[i] +dt * (-g/l *np.sin(theta_mid))

        # t_mid = t[i] + 0.5 * dt


    return t, theta, dt, max_theta





list = [20, 40, 80, 160, 320]

print("dt | max(0*) | max(0*) - 0*")

for i in list:
    t, theta, dt, max_theta = pen(i)
    plt.plot(t, theta, label=f"n = {i}")
    print(f"{dt} | {max_theta} | {max_theta - np.pi/3}")



plt.xlabel('t')
plt.ylabel('theta')
plt.show()
pen(200)
```
